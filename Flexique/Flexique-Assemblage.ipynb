# Lire Lexique3

```python
lexique3=pd.read_csv(filepath_or_buffer=lexique3Rep+"lexique381.txt",sep="\t",encoding="utf8")
```

```python
lexique3.columns
```

### Garder seulement les 11 premières colonnes

```python
lex3=lexique3.drop(lexique3.columns[10:],axis=1)
```

### Garder seulement les adjectifs
- renommer les colonnes
- ajouter une fréquence moyenne entre Film et Livre

```python
alex3=lex3[lex3["4_cgram"]=="ADJ"][:]
alex3.columns=["ortho","phon","lexeme","cgram","genre","nombre","filmLemme","livreLemme","filmForme","livreForme"]
alex3["freq"]=(alex3["filmForme"]+alex3["livreForme"])/2
```

```python
alex3
```

# Incorporer Lexique3 à Flexique

### Pivoter la table au format de Flexique
- ajouter des proportions pour chaque case

```python
palex3=pd.pivot_table(alex3,index="lexeme",values="freq",columns=["genre","nombre"],fill_value=0)
```

```python
palex3.columns = palex3.columns.to_series().str.join('')
```

```python
palex3=palex3[[g+n for g in "mf" for n in "sp"]][:]
```

```python
palex3["freqLexeme"]=palex3["ms"]+palex3["mp"]+palex3["fs"]+palex3["fp"]
```

```python
for c in [g+n for g in "mf" for n in "sp"]:
    palex3[c+"Prop"]=palex3[c]/palex3["freqLexeme"]
```

```python
palex3
```

### Faire la jointure entre Flexique et Palex3

```python
flexique.join(palex3)
```

```python
flex3=flex3[["lexeme","case","forme","freq"]][:]
flex3["freq"]=flex3["freq"].fillna(0)
```
