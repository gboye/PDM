# Transforme un paradigme CSV en entrée PKL pour Swim

```python
import pandas as pd
import pickle
```

```python
repPrefix="/Volumes/gilles/Transfert/Copies-iMac-GB/2015-Data/"
filePrefix=repPrefix+"StemSpaces/LongitudinalesRnd/"
csvParadigme=filePrefix+"Longitudinal-00-T1000000-F29754-X-paradigmes-Swim2.csv"
```

```python
paradigmes=pd.read_csv(csvParadigme,sep=";",encoding="utf8")
if u"Unnamed: 0" in paradigmes:
    del paradigmes[u"Unnamed: 0"]
```

```python
paradigmes
```

```python
sample=paradigmes.melt(id_vars="lexeme")
sample.columns=["lexeme","case","phono"]
sample
```

```python
sample["tir1"]=1*sample["phono"].notnull()
sample[sample["tir1"]>0]
```

```python
with open(csvParadigme.replace(".csv",".pkl"), 'wb') as outFile:
    pickle.dump(sample[sample["tir1"]>0], outFile, pickle.HIGHEST_PROTOCOL)
```

```python
with open(csvParadigme.replace(".csv",".pkl"), 'rb') as inFile:
    test=pickle.load(inFile)
```
