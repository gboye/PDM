```python
import csv,pickle,codecs
import pandas as pd
```

```python
lexiquePrefix="2015-Data/MGC-160104-basesDerives"
with open('2015-Data/MGC-160104-Derives2.pkl', 'rb') as input:
    derives = pickle.load(input)
```

```python
baseDerives=derives.set_index("lexeme")["base"].to_dict()
baseDerives
```

```python
verbesBRASS=pd.read_csv(lexiquePrefix+"-01-N-paradigmes.csv",sep=";",encoding="utf8")
del verbesBRASS[u"Unnamed: 0"]
verbesBRASS
```

```python
verbesSILVER=pd.read_csv(lexiquePrefix+"-01-N-SILVER.csv",sep=";",encoding="utf8")
del verbesSILVER[u"Unnamed: 0"]
#del verbesSILVER[u"score"]
#del verbesSILVER[u"ajouts"]

verbesSILVER
```

```python
verbesGOLD=pd.read_csv(lexiquePrefix+"-01-N-GOLD.csv",sep=";",encoding="utf8")
del verbesGOLD[u"Unnamed: 0"]
verbesGOLD
```

```python
derivesBRASS=pd.read_csv(lexiquePrefix+"-derives-01-N-paradigmes.csv",sep=";",encoding="utf8")
del derivesBRASS[u"Unnamed: 0"]
derivesBRASS
```

```python
derivesSILVER=pd.read_csv(lexiquePrefix+"-derives-01-N-Silver.csv",sep=";",encoding="utf8")
del derivesSILVER[u"Unnamed: 0"]
del derivesSILVER[u"score"]
del derivesSILVER[u"ajouts"]
derivesSILVER
```

```python
derivesGOLD=pd.read_csv(lexiquePrefix+"-derives-01-N-Gold.csv",sep=";",encoding="utf8")
del derivesGOLD[u"Unnamed: 0"]
derivesGOLD
```

```python
listVerbesBRASS=verbesBRASS["lexeme"].tolist()
listVerbesSILVER=verbesSILVER["lexeme"].tolist()
listVerbesGOLD=verbesGOLD["lexeme"].tolist()
for derive in baseDerives:
    if baseDerives[derive] in listVerbesBRASS:
        if derivesBRASS[derivesBRASS["lexeme"]==derive]["FS"].values:
            verbesBRASS.loc[verbesBRASS["lexeme"]==baseDerives[derive],"FS"]=derivesBRASS[derivesBRASS["lexeme"]==derive]["FS"].values[0]
        if derivesBRASS[derivesBRASS["lexeme"]==derive]["FP"].values:
            verbesBRASS.loc[verbesBRASS["lexeme"]==baseDerives[derive],"FP"]=derivesBRASS[derivesBRASS["lexeme"]==derive]["FP"].values[0]
    if baseDerives[derive] in listVerbesSILVER:
        if derivesSILVER[derivesSILVER["lexeme"]==derive]["FS"].values:
            verbesSILVER.loc[verbesSILVER["lexeme"]==baseDerives[derive],"FS"]=derivesSILVER[derivesSILVER["lexeme"]==derive]["FS"].values[0]
        if derivesSILVER[derivesSILVER["lexeme"]==derive]["FP"].values:
            verbesSILVER.loc[verbesSILVER["lexeme"]==baseDerives[derive],"FP"]=derivesSILVER[derivesSILVER["lexeme"]==derive]["FP"].values[0]
    if baseDerives[derive] in listVerbesGOLD:
        if derivesGOLD[derivesGOLD["lexeme"]==derive]["FS"].values:
            verbesGOLD.loc[verbesGOLD["lexeme"]==baseDerives[derive],"FS"]=derivesGOLD[derivesGOLD["lexeme"]==derive]["FS"].values[0]
        if derivesGOLD[derivesGOLD["lexeme"]==derive]["FP"].values:
            verbesGOLD.loc[verbesGOLD["lexeme"]==baseDerives[derive],"FP"]=derivesGOLD[derivesGOLD["lexeme"]==derive]["FP"].values[0]
```
