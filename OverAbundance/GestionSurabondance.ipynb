```python
import pandas as pd
import random as rd
import itertools as it
```

```python
tam=[t+p for t in ["pi","ii","ps","pc"] for p in [str(num)+nb for num in range(1,4) for nb in "SP"]]
```

```python
def tirageTAM():
    return [rd.choice(tam) for i in range(rd.randint(1,10))]
```

```python
pseudoLexicon={}
for lexIndex in range(2000):
    pseudoLexicon[lexIndex]={}
    for case in tirageTAM():
        if not case in pseudoLexicon[lexIndex]:
            pseudoLexicon[lexIndex][case]=[]
        pseudoLexicon[lexIndex][case].append(rd.randint(1,1000))
```

```python
paradigmes=pd.DataFrame.from_dict(pseudoLexicon,orient='index')
```

```python
def subParadigm(c1,c2):
    return paradigmes[paradigmes[c1].notnull() & paradigmes[c2].notnull()][[c1,c2]]
```

```python
paire=subParadigm("pi3S","pi1P")
```

```python
for c1,c2 in it.combinations(tam,2):
    paire=subParadigm(c1,c2)
    doublets=paire[(paire[c1].str.len()>1) & (paire[c2].str.len()>1)]
    if len(doublets)>0:
        print doublets
```
