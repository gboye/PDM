```python
def multireplace(dic,text,greedy=True):
    
    if greedy:
        keys = sorted(dic,key=lambda x:-len(x))
    else:
        keys = sorted(dic,key=lambda x:len(x))

    
    pattern = re.compile("|".join(re.escape(k) for k in keys))    
    result = pattern.sub(lambda m: dic[m.group(0)], text)
    return result

    
```

# Stress

Dunstan Brown:

> if no stress is marked then it will be on the first syllable.

```python
HTML(translation_guide.iloc[3:5,:].to_html())
```

```python
accent_on_vowel = {"a":"á","e":"é","i":"í","o":"ó","u":"ú"}

        
def new_stress(cell,i):
    return cell[:i] + accent_on_vowel[cell[i]] + cell[i+1:]

def attribute_stress(cell):
    if cell != "#DEF#":
        vow_before = "[aeiou][^a^e^i^o^u]*?"
        if '@"' in cell :
            regex = re.compile(vow_before +"@\"")
        elif "-&" in cell :
            regex = re.compile(vow_before +"-&")
        else:
            regex = re.compile("[aeiou]")

        i = regex.search(cell).start()
        cell = multireplace({"-&":"",'@"':""},cell)
        return new_stress(cell,i)
    return cell
```

```python
data[trans_cols] = data[trans_cols].applymap(attribute_stress)
```

# Palatals, affricates

```python
HTML(translation_guide.iloc[5:15,:].to_html())
```

```python
palatal_mark = {"'":"ʲ", "@'":"ʲ"}

palatal_1 = {
             
             # ch, shch and c transcription. Can't be palatalized. 
             "ch":"ʨ",    "chʲ":"ʨ",
             "shch":"ɕː", "shchʲ":"ɕː",
             "c":"ʦ",     "cʲ":"ʦ",
             
             # j can't be palatalized.
             "jʲ":"j",
             
            # Final e is pronounced "i"
             "ije$":"iji",
             
            # i palatalization of velars (even when accented)
            "ki":"kʲi",
            "gi":"gʲi",
            "xi":"xʲi",
            "kí":"kʲí",
            "gí":"gʲí",
            "xí":"xʲí",
             
             # Simple translations
            "zh":"ʒ",
            "sh":"ʃ"}

def others(cell):
    # Palatalization of cons before e (only if the consonant is palatalizable)
    cell = re.compile("(?<![ʲːʨjʒʃʦaáeéiíoóuú])e").sub("ʲe",cell)
    cell = re.compile("(?<![ʲːʨjʒʃʦaáeéiíoóuú])é").sub("ʲé",cell)
    
    # o becomes e after a palatal.
    cell = re.compile("(?<=[ʲːʨj])o").sub("e",cell)
    return cell

data[trans_cols] = data[trans_cols].applymap(lambda x:multireplace(palatal_mark,x))
data[trans_cols] = data[trans_cols].applymap(lambda x:multireplace(palatal_1,x))
data[trans_cols] = data[trans_cols].applymap(others)
```
