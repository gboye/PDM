### Lire l'extrait de BDLexique des dérivés

```python
with open(lexiquePrefix+'-Derives2.pkl', 'rb') as input:
    derives = pickle.load(input)

bases=set(derives["base"].tolist())
```

### Définir le dialecte de l'échantillon

```python
echantillonPrefix+=phonologicalMap
if debug: print(phonologicalMap)
neutralisationsNORD=(u"6û",u"9ê")
neutralisationsSUD=(u"e2o",u"E9O")
if phonologicalMap=="-N":
    neutralisations=neutralisationsNORD
elif phonologicalMap=="-S":
    neutralisations=neutralisationsSUD
else:
    neutralisations=(u"",u"")
    phonologicalMap=("-X")
bdlexiqueIn = unicode(u"èò"+neutralisations[0])
bdlexiqueNum = [ord(char) for char in bdlexiqueIn]
neutreOut = unicode(u"EO"+neutralisations[1])
neutralise = dict(zip(bdlexiqueNum, neutreOut))
```

```python
def recoder(chaine,table=neutralise):
    if type(chaine)==str:
        temp=unicode(chaine.decode('utf8')).translate(table)
        result=temp.encode('utf8')
    elif type(chaine)==unicode:
        result=chaine.translate(table)
    return result
```

### Faire la liste des cases de l'échantillon 01

```python
corpusApprentissage=pd.read_csv(nomCorpusApprentissage,sep=";",encoding="utf8")
del corpusApprentissage[u"Unnamed: 0"]
corpusApprentissage=corpusApprentissage.dropna(axis=1,how='all')
sampleCases=corpusApprentissage.columns.tolist()
```

### Lire l'extrait de BDLexique des formes verbales
- lire les formes verbales
- recoder en fonction du dialecte

```python
with open(lexiquePrefix+'-Verbes2.pkl', 'rb') as input:
    verbes = pickle.load(input)

verbes["phono"]=verbes["phono"].apply(lambda x: recoder(x))
verbes
```

```python
derivables=verbes[(verbes["lexeme"].isin(bases)) & (verbes["case"].isin(paradigmeCases))]
derivables[derivables["lexeme"]=="surtondre"]
```

```python
formesBases=derivables.reset_index()
del formesBases["index"]
formesBases
```

```python
basesManquantes=pd.read_csv(nomManquants,sep=";",encoding="utf8")
basesManquantes["ms"]=basesManquantes["ms"].fillna("")
basesManquantes["case"]=basesManquantes["vs"]+basesManquantes["ms"]
formesManquantes=basesManquantes[["lexeme","phono","case"]]
formesManquantes
```

```python
formesManquantes.loc[:,"phono"]=formesManquantes.loc[:,"phono"].apply(lambda x: recoder(x))
formesManquantes.loc[:,"case"]=formesManquantes.loc[:,'case'].astype(unicode) #= formesManquantes['case'].map(lambda x: x.decode("utf8"))
formesManquantes
#basesManquantes
```

```python
formesBases=formesBases.append(formesManquantes, ignore_index=True)
set(formesBases["case"].tolist())
```

```python
formesBases[formesBases["lexeme"]=="saillir"]
```

```python
paradigmes=pd.pivot_table(formesBases, values='phono', index=['lexeme'], columns=['case'], aggfunc=lambda x: ",".join(x)).reset_index().reindex()
```
