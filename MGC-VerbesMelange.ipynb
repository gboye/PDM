# Lecture du tableau de VERBES original

```python
verbes=pd.read_pickle(path=nomRepertoire+filePrefix+'-Verbes3.pkl')
```

# Ajouter une colonne pour la probabilité du lexème

```python
lexemeProb=verbes.groupby("lexeme")["prob"].sum()
```

```python
def probLexeme(lexeme):
    result=lexemeProb[lexeme]
    return result
```

```python
verbes["probLexeme"]=verbes["lexeme"].apply(probLexeme)
```

# Ajouter une colonne pour la proportion de la forme au sein du lexème

```python
verbes["factForme"]=verbes["prob"]/verbes["probLexeme"]
```

# Faire une table de correspondance entre lexèmes au hasard

```python
vOriginal=lexemeProb.index.tolist()
vRandom=lexemeProb.index.tolist()
seed(200363)
shuffle(vRandom)
dictRndVerbes={vOriginal[i]:vRandom[i] for i in range(len(vOriginal))}
```

```python
mapVerbes=[]
for l in u"être avoir faire aller parler".split(" "):
    mapVerbes.append(dictRndVerbes[l])
print " ".join(mapVerbes)
```

# Associer une nouvelle fréquence de lexème d'après la table de correspondance

```python
def corrLexeme(lexeme):
    result=dictRndVerbes[lexeme]
    return result
```

```python
verbes["baseLexeme"]=verbes["lexeme"].apply(corrLexeme)
```

```python
verbes["baseLexemeProb"]=verbes["baseLexeme"].apply(probLexeme)
```

```python
verbes["corrProb"]=verbes["baseLexemeProb"]*verbes["factForme"]
```

```python
ajustementFreq=int(10/verbes["corrProb"].min()*verbes["freq"].min())
```

```python
ajustementFreq=verbes["freq"].sum()
```

```python
verbes["corrFreq"]=verbes["corrProb"]*ajustementFreq
```

```python
verbes["corrFreq"].min()
```

# Refaire la table au format original avec les nouvelles colonnes

```python
verbes.rename(columns={"freq":"origFreq","prob":"origProb"},inplace=True)
verbes.rename(columns={"corrFreq":"freq","corrProb":"prob"},inplace=True)
```

```python
verbes.sort_values("freq",ascending=False)
```
