```python
import pandas as pd
import pickle
from __future__ import print_function
```

```python
nomSortie="2015-Data/MGC-160104-01-N-paradigmes-ext-Silver.csv"
nomDerives="2015-Data/MGC-160104-Derives2.pkl"
```

```python
sortie=pd.read_csv(nomSortie,sep=";",encoding="utf8")
del sortie["Unnamed: 0"]
```

```python
with open(nomDerives, 'rb') as input:
    derives = pickle.load(input)
basesDerives=set(derives["base"].tolist())
```

```python
for row in sortie[sortie["lexeme"].isin(basesDerives)].iterrows():
    for element in row[1]:
        if type(element)==unicode:
            print (element,end=", ")
    print ()
    print ()
```
