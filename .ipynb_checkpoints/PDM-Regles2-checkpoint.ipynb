```python
import re
import sys
import copy
```

```python
sepPaire="=>"
paireCase=""
pairesCases=[]
pairesCasesRegles={}
sortiesCases={}
Classes={}
classeDict={}
classeListe=[]
reglesSPE={}
reglesPDM=[]
```

```python
def paireEntree(paire):
    return paire.split("=>")[0]
def paireSortie(paire):
    return paire.split("=>")[1]
def fairePaire(entree,sortie):
    return entree+sepPaire+sortie
```

```python
def formaterRegle(line):
    m=re.search("^(\S+)\s+-->\s+(\S+)\s+/\s+(\S+)\s+___\s+(\S+)\s+(.*)$",line)
    if m:
        a=m.group(1)
        if a=="[]":
            a=""
        b=m.group(2)
        if b=="[]":
            b=""
        c="".join(m.group(3).split(","))
        if c[0]=="X":
            c="^(.*"+c[1:]+")"
        if c[0]=="#":
            c="^("+c[1:]+")"
        d=m.group(4)
        if d[-1]=="X":
            d="("+d[:-1]+".*)$"
        if d[-1]=="#":
            d="("+d[:-1]+")$"
        reste=m.group(5)
        return (c+a+d,b,reste)
    else:
        print ("NOT A RULE")
        return
```

###Ajouter une règle à la liste et renvoyer le numéro de la règle

si la règle existe déjà, on renvoie simplement le numéro existant

```python
def ajouterRegle(regle):
    (regex,subst)=regle
    if not regle in reglesPDM:
        reglesPDM.append(regle)
    return reglesPDM.index(regle)
```

###Interface avec la sortie de PredSPE

```python
def lirePredSPE(nomFichier):
    try:
        fichierPredSPE=open(nomFichier,"r")
    except IOError:
        print ('I could not open the file', nomFichier)
        sys.exit()
    zoneRegles=False
    zoneClasses=False
    zoneClasse=False
    for line in fichierPredSPE:
        line=line.strip()
        m=re.search("^\s*(\S+)\s+==>\s+(\S+)\s*$",line)
        if m:
            paireCase=m.group(1)+sepPaire+m.group(2)
#            print (paireCase)
            reglesSPE[paireCase]=[]
            pairesCases.append(paireCase)
            case=paireSortie(paireCase)
            if not case in sortiesCases:
                sortiesCases[case]=[]
            sortiesCases[case].append(paireEntree(paireCase))
        if "*****" in line:
            zoneRegles=False
            zoneClasses=False
        if "done." in line:
            zoneRegles=False
        if "conditional entropy for" in line:
            zoneClasses=False        
            paireCase=""
        if zoneRegles:
#            print ("Règle",line)
#            reglesSPE[paireCase].append(line)
            (inPatron,mod,reste)=formaterRegle(line)
            if not paireCase in pairesCasesRegles:
                pairesCasesRegles[paireCase]=[]
            pairesCasesRegles[paireCase].append((ajouterRegle((inPatron,mod))))
        if zoneClasses:
            if "local conditional entropy" in line:
                zoneClasse=False
                nomClasse="-".join(sorted(classeListe,key=int))
                if not paireCase in Classes:
                    Classes[paireCase]={}
                Classes[paireCase][nomClasse]=classeDict
            if zoneClasse:
                (inPatron, mod, reste)=formaterRegle(line)
                n=re.search("^.*—(.+)%.*$",reste)
                if n:
                    classeDict[reglesPDM.index((inPatron,mod))]=float(n.group(1))
                    classeListe.append(str(reglesPDM.index((inPatron,mod))))
#                    print (reglesPDM.index((inPatron,mod)),float(n.group(1)))
                else:
                    print ("NO DISTRIBUTION")
            if "class" in line:
                zoneClasse=True
                classeDict={}
                classeListe=[]
        if "Inferring rules" in line:
            zoneRegles=True
        if "done." in line:
            zoneClasses=True
```

##Génération de distributions à partir d'une forme

- Eventuellement éliminer le tri des numéros de règle pour gagner du temps

```python
def distributionVecteurs(matrice):
    formes={}
    total=0
    for vecteur in matrice:
        for fc in vecteur:
            forme=fc.f()
            coef=fc.c()
            if not forme in formes:
                formes[forme]=0
            formes[forme]+=coef
            total+=coef
    distribution=[]
    supprime=False
    for forme in formes:
        coef=float("{0:.2f}".format(formes[forme]/total))
        if coef>seuilDistribution:
            distribution.append(FormeCoef(forme,coef))
        else:
            supprime=True
    if supprime==True:
        distribution=distributionVecteurs([distribution])
    return sorted(distribution,key=lambda couple: couple[1],reverse=True)
```

```python
def transformerForme(inPatron,mod,forme):
    extrait=re.search(inPatron,forme)
    if extrait:
        sortie=extrait.group(1)+mod+extrait.group(2)
        return sortie
    else:
        return False

def outputFormes(paireCase,FormeCoef):
    lines=[]
    numsClasse={}
    (forme,part)=FormeCoef
    for numero in pairesCasesRegles[paireCase]:
        (inPatron,mod)=reglesPDM[numero]
#        extrait=re.search(inPatron,forme)
#        if extrait:
#            sortie=extrait.group(1)+mod+extrait.group(2)
        sortie=transformerForme(inPatron,mod,forme)
        if sortie:
            numsClasse[numero]=sortie
    alphaKeys=[]
    for num in sorted(numsClasse.keys()):
        alphaKeys.append(str(num))
    classe="-".join(alphaKeys)
    if not classe in Classes[paireCase]:
        print ("FORME SANS SORTIE",forme)
    else:
        for numero in numsClasse:
            if Classes[paireCase][classe][numero]>0:
                lines.append((numsClasse[numero],Classes[paireCase][classe][numero]*part))
    return distributionVecteurs([lines])  
```

```python
class FormeCoef:
    def __init__(self,forme,coef):
        self.forme=forme
        self.coef=coef
    def __getitem__(self,index):
        if index==0:
            return self.forme
        elif index==1:
            return self.coef
        else:
            raise ValueError("out of range [0,1]")
    def __setitem__(self,index,value):
        if index==0:
            self.forme=value
        elif index==1:
            self.coef=value
        else:
            raise ValueError("out of range [0,1]")
    def f(self):
        return self.forme
    def c(self):
        return self.coef
    def val(self):
        return (self.forme,self.coef)
    
class Case:
    def __init__(self,name=""):
        self.name=name
        self.valeurs=[]
        self.coefficient={}
        self.normalValeurs=[]
        self.normalCoefficient={}
    def nom(self):
        return self.name
    def ajouterFormeCoef(self,FormeCoef):
        self.valeurs.append(FormeCoef)
        if not FormeCoef.f() in self.coefficient:
            self.coefficient[FormeCoef.f()]=0
        self.coefficient[FormeCoef.f()]+=FormeCoef.c()
        self.normalValeurs=distributionVecteurs([self.valeurs])
        for fc in self.normalValeurs:
            self.normalCoefficient[fc.f()]=fc.c()
    def ajouterFormesCoefs(self,FormesCoefs):
        for FormeCoef in FormesCoefs:
            self.ajouterFormeCoef(self,FormeCoef)
    def contenu(self,original=False):
        if original:
            return self.valeurs
        else:
            return self.normalValeurs
    def coef(self,forme,original=False):
        if original:
            if forme in self.coefficient:
                return self.coefficient[forme]
            else:
                return 0
        else:
            if forme in self.normalCoefficient:
                return self.normalCoefficient[forme]
            else:
                return 0
    def coefficients(self,original=False):
        if original:
            return self.coefficient
        else:
            return self.normalCoefficient
```
