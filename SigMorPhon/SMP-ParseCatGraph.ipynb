```python
pdTrain=pd.read_csv(filepath_or_buffer=trainingDataFile,sep="\t",header=None,encoding="utf8")
```

```python
def normalizeTag(tag):
    parts=tag.split(",")
    result=[]
    for part in parts:
        m=re.match("^[^=]*=(.*)$",part)
        if m:
            result.append(m.group(1))
    return "x".join(result)
```

```python
pdTrain.columns=["t1","f1","t2","f2"]
pdTrain["tag1"]=pdTrain["t1"].apply(normalizeTag)
pdTrain["tag2"]=pdTrain["t2"].apply(normalizeTag)
pdTrain["p1"]=pdTrain["f1"]+"-"+pdTrain["tag1"]
pdTrain["p2"]=pdTrain["f2"]+"-"+pdTrain["tag2"]
```

```python
def splitTag(cumul):
    result=[]
    if len(cumul)==1:
        m=re.match(ur"(.*){(.+)\/(.+)}(.*)",cumul[0])
        if m:
            result.append(m.group(1)+m.group(2)+m.group(4))
            result.append(m.group(1)+m.group(3)+m.group(4))
            result=splitTag(result)
        else:
            return cumul
    else:
        for element in cumul:
            result.extend(splitTag([element]))
    return result  
```

```python
def splitTags(df,colonne):
    '''
    Calcul d'une dataframe sans Tag ambigu par dédoublement des valeurs
    '''
    test=df.reset_index()
    del test["index"]
    splitIndexes=[]
    for index,ligne in test.iterrows():
        valeurs=splitTag([ligne[colonne]])
        if len(valeurs)>1:
            nouvelleLigne=ligne
            for valeur in valeurs:
                nouvelleLigne[colonne]=valeur
                test=test.append(nouvelleLigne,ignore_index=True)
            splitIndexes.append(index)
    if splitIndexes:
        test=test.drop(test.index[splitIndexes])
    return test
```

```python
pdLex=splitTags(pdLex,"tag1")
```

```python
german1graphe=(u"üöäwc",u"y2Evk")
german2=[u"sch",u"ch",u"sc",u"x",u"ß",u"ck",u"#v",u"#s",u"uç",u"oç",u"aç",u"e#",u"qu",u"ci",u"ce"]
bdlexique2=[u"S",u"ç",u"sk",u"ks",u"ss",u"k",u"#f",u"#z",u"ux",u"ox",u"ax",u"6#",u"ku",u"si",u"se"]
germanIn = unicode(german1graphe[0])
germanNum = [ord(char) for char in germanIn]
bdlexiqueOut = unicode(german1graphe[1])
neutralise = dict(zip(germanNum, bdlexiqueOut))
```

```python
def recoder(chaine):
    table=neutralise
    if type(chaine)==str:
        temp="#"+unicode(chaine.decode('utf8').lower())+"#"
        for numBigramme,bigramme in enumerate(german2):
            temp=temp.replace(bigramme,bdlexique2[numBigramme])
        temp.translate(table)
        result=temp.strip("#").encode('utf8')
        print result
    elif type(chaine)==unicode:
        chaine="#"+chaine.lower()+"#"
        for numBigramme,bigramme in enumerate(german2):
            chaine=chaine.replace(bigramme,bdlexique2[numBigramme])
        result=chaine.strip("#")
        result=result.translate(table)
    else:
        result=chaine
    return result
```

```python
pdLex["phono"]=pdLex["f1"].apply(recoder)
```
